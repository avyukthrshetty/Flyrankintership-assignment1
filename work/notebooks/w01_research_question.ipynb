## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The starter dataset contains 30,000 pages and 44 features describing SEO performance. The correlation between search volume and impressions is only 0.001, showing that search volume alone is not a strong predictor of page impressions. In addition, 54.21% of the pages are marked as declining, suggesting there are many pages that could potentially benefit from content refresh. These observations support the need for a data-driven prioritization approach.


```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

corr = df["search_volume"].corr(df["impressions_90d"])
print("Correlation between Search Volume and Impressions:", round(corr, 3))

declining = (df["trend_direction"] == "down").mean()
print("Percentage of Declining Pages:", round(declining * 100, 2), "%")
```
